### Criando uma nova coluna que contém os valores da CDU 

> "Os itens do acervo em uma biblioteca são organizados por um sistema de 
classificação de acordo com o respectivo tema. Existem diversos sistemas, <br>
mas este conjunto está de acordo com a CDU - Classificação Decimal Universal.

```python
emprestimos['CDU'] = emprestimos['localizacao'].replace({valor: 'Generalidades' if valor < 100 else 
                                                                             'filosofia e psicologia' if valor < 200 else
                                                                             'religiao' if valor < 300 else 
                                                                             'Ciencias socias' if valor < 400 else
                                                                             'classe vaga' if valor < 500 else 
                                                                             'matemática e ciências naturais' if valor < 600 else 
                                                                             'ciências aplicadas' if valor < 700 else 
                                                                             'belas artes' if valor < 800 else 
                                                                             'linguagem' if valor < 900 else 
                                                                             'Geografia. Biografia. História.'
                                                                             
                                                                             for valor in emprestimos['localizacao']})
```

```python
def destribuirFrequencia(x):
    '''
    Essa função pega as variaveis do dataset e realiza a distribuição de frequencias 
    x = variável categorica do dataset emprestimos
    '''
    distribuicao = pd.DataFrame()
    distribuicao['Quantidade'] = emprestimos[x].value_counts()
    distribuicao['Frequência'] = round(emprestimos[x].value_counts(normalize = True) * 100, 1)
    
    return distribuicao.reset_index()
    
```
